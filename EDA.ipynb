## Read in the data

```python
# import the data 

import pandas as pd

df = pd.read_csv('../Data/happiness_survey_data.csv')
df.head()
```

```python
# there are 2089 rows and 6 columns of data
df.shape
```

```python
# the years range from 2005 - 2021
# happiness scores range from 2 - 8
df.describe()
```

## Explore the data

```python
# Filter out any data before 2010 and after 2019
df = df[(df['year'] >= 2010) & (df['year'] <= 2019)]
```

```python
# Group the data by country and calculate the maximum happiness score for each one
df_grouped = df.groupby('country_name')['happiness_score'].max().reset_index()
df_grouped.head()
```

```python
# Sort the grouped countries by happiness score and return the top five
summed_happiness = df_grouped.sort_values('happiness_score', ascending = False).head(5)
```

```python
# Group the data by country and calculate the average happiness score for each one
(df.groupby('country_name')['happiness_score'].mean()).reset_index()
```

```python
# Sort the grouped countries by happiness score and return the top five
avg_happiness = (df.groupby('country_name')['happiness_score'].mean()).reset_index().sort_values('happiness_score', ascending = False).head(5)
```
